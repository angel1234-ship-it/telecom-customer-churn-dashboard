```python
import pandas as pd
df=pd.read_csv(r"E:\DATA ANALYST\PROJECT\DEFORESTATION\Telecom\telecom_customer_churn.csv")
df
```

```python
dupli=df.duplicated()
print(dupli)
```

```python
df['Churn Category']=df['Churn Category'].fillna('Active')
df
```

```python
df['Churn Reason']=df['Churn Reason'].fillna('Not Applicable')
df
```

```python
pd.set_option('display.max_columns',40)
df
```

```python
df.tail(10)
```

```python
df['Offer']=df['Offer'].fillna('None')
df
```

```python
place=df['Avg Monthly Long Distance Charges'].mean()
print(place)
```

```python
df['Avg Monthly Long Distance Charges']=df['Avg Monthly Long Distance Charges'].fillna(place)
df
```

```python
df['Multiple Lines']=df['Multiple Lines'].fillna('No')
df
```

```python
df.loc[df['Monthly Charge'] < 0,'Monthly Charge']=0
df
```

```python
negative_charges = df[df['Monthly Charge'] < 0]
print(negative_charges)
```

```python
bins = [0, 12, 24, float('inf')]
labels = ['New', 'Established', 'Loyal']
df['Tenure_Category'] = pd.cut(df['Tenure in Months'], bins=bins, labels=labels, right=False)
df
```

```python
df.to_csv('cleanedtelecom_data.csv', index=False)
```
