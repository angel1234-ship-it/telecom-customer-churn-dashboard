```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

# --- 1. Create a Jittered Y-Axis for Visualization Only ---
# np.random.uniform creates random numbers between -0.15 and 0.15
# We add this random noise to the discrete Competitor_Vulnerability_Score
df['Y_Jittered'] = df['Competitor_Vulnerability_Score'] + np.random.uniform(-0.15, 0.15, size=len(df))

# --- 2. GENERATE THE IMPROVED SCATTER PLOT ---

plt.figure(figsize=(18, 8))

sns.scatterplot(
    x=df['Total Revenue'],
    y=df['Y_Jittered'], # Use the jittered Y-axis
    hue=df['Customer Status'],
    palette={'Stayed': 'green', 'Churned': 'red', 'Joined': 'blue'},
    size=df['Avg Monthly GB Download'],
    sizes=(20, 300), 
    alpha=0.6 
)

# --- Add Strategic Quadrant Markers ---
revenue_threshold = df['Total Revenue'].quantile(0.75)
plt.axvline(x=revenue_threshold, color='orange', linestyle='--', label='Top 25% Revenue')
plt.axhline(y=2, color='red', linestyle=':', label='High Vulnerability Threshold')

# --- Final Plot Aesthetics ---
plt.title('Customer Vulnerability vs. Revenue (Targeting the Top-Right Quadrant)', fontsize=16)
plt.xlabel('Customer Value: Total Revenue ($)', fontsize=14)
plt.ylabel('Competitor Vulnerability Score (Commitment vs. Premium Support)', fontsize=14)

# Keep the original Y-tick labels, as the jitter is small and conceptual
plt.yticks([0, 1, 2, 3], ['Lowest Vulnerability', 'Low/Medium', 'High Vulnerability (M-t-M)', 'Max Vulnerability'])

plt.legend(title='Customer Status & Thresholds', loc='upper right')
plt.grid(True, which='major', linestyle='--', linewidth=0.5, axis='y') # Only show horizontal grid for clarity

plt.show()
```

```python
# --- 2. Feature Engineering: Total Revenue (X-axis) ---
# Convert essential columns to numeric, filling errors with 0
numeric_cols = ['Total Charges', 'Total Long Distance Charges', 'Total Extra Data Charges', 'Total Refunds', 'Monthly Charge']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Calculate Total Revenue
df['Total Revenue'] = df['Total Charges'] + df['Total Long Distance Charges'] + df['Total Extra Data Charges'] - df['Total Refunds']


# --- 3. Feature Engineering: Competitor Vulnerability Score (Y-axis) ---
# Map Premium Tech Support to vulnerability (1=No Support, 0=Yes)
df['Support_Vulnerability'] = df['Premium Tech Support'].apply(lambda x: 1 if x == 'No' else 0)

# Map Contract Type to vulnerability (2=M-t-M, 1=1yr, 0=2yr)
contract_map = {'Two Year': 0, 'One Year': 1, 'Month-to-Month': 2}
# Use .map() and fill any missing contract values with the max risk (2)
df['Contract_Vulnerability'] = df['Contract'].map(contract_map).fillna(2)

# Combined Vulnerability Score (Range 0 to 3)
df['Competitor_Vulnerability_Score'] = df['Support_Vulnerability'] + df['Contract_Vulnerability']


# --- 4. Define the Strategic Target Quadrant Thresholds ---
# Revenue Threshold: Top 25% of all customers by Total Revenue
revenue_threshold = df['Total Revenue'].quantile(0.75)

# Vulnerability Threshold: Score >= 2 (High Vulnerability threshold)
vulnerability_threshold = 2


# --- 5. Filter the Target Segment (Green Dots in Top-Right Quadrant) ---
target_segment = df[
    (df['Customer Status'] == 'Stayed') &                                # Only 'Stayed' customers (Green dots)
    (df['Total Revenue'] >= revenue_threshold) &                        # Only High Value (Right of orange line)
    (df['Competitor_Vulnerability_Score'] >= vulnerability_threshold)   # Only High Vulnerability (Above red line)
]


# --- 6. Calculate Key Financial Metrics ---
customer_count = len(target_segment)
total_revenue_at_risk = target_segment['Total Revenue'].sum()
annual_revenue_at_risk = target_segment['Monthly Charge'].sum() * 12 # Approximation
average_monthly_charge = target_segment['Monthly Charge'].mean()


# --- 7. Print Results for the Recommendation ---
print("\n--- Financial Metrics for High-Risk, High-Value Segment ---")
print(f"Number of Stayed Customers in Target Quadrant: {customer_count:,}")
print(f"Total Cumulative Revenue (LTV) at Risk: ${total_revenue_at_risk:,.2f}")
print(f"Estimated Annual Recurring Revenue (ARR) at Risk: ${annual_revenue_at_risk:,.2f}")
print(f"Average Monthly Charge in Target Segment: ${average_monthly_charge:,.2f}")
print(f"\nStrategic Thresholds:")
print(f"  Revenue >= ${revenue_threshold:,.2f}")
print(f"  Vulnerability Score >= {vulnerability_threshold}")
```

```python

# --- 5. Analyze and Visualize the 'Offer' Distribution within the Target Segment ---

# Calculate the distribution (in percentages)
offer_distribution = target_segment['Offer'].value_counts(normalize=True).mul(100).round(2).sort_values(ascending=False)

print("\n--- Offer Distribution in the 464 High-Risk, High-Value Customers ---")
print(offer_distribution.to_string())

# Create a bar chart for visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=offer_distribution.index, y=offer_distribution.values, palette="viridis")

plt.title('Distribution of Existing Offers for High-Risk, High-Value Customers (N=464)', fontsize=14)
plt.xlabel('Offer Type', fontsize=12)
plt.ylabel('Percentage of Segment (%)', fontsize=12)
plt.yticks(range(0, int(offer_distribution.max()) + 10, 10))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of the bars
for index, value in enumerate(offer_distribution.values):
    plt.text(index, value + 0.5, f'{value:.2f}%', ha='center')

plt.savefig('offer_distribution_target_segment.png')
plt.show() # Use plt.show() to display in a notebook environment
```
